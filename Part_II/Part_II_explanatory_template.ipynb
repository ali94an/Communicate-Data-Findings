## Dataset Overview and Executive Summary

The dataset contains information about bike-sharing trips in February 2019. Key variables include:
- **Trip Duration**: Measured in seconds, indicating the time spent on each trip.
- **User Type**: Differentiates between customers and subscribers.
- **Member Gender**: Provides gender information for users.
- **Member Birth Year**: Indicates the year of birth for each user.

**Summary of Findings**:
- **Duration Patterns**: Customers generally have longer trips compared to subscribers.
- **Demographics**: The majority of users are male, with female and other genders representing smaller proportions.
- **Age Influence**: Younger users tend to take longer trips, with trip durations decreasing as birth year increases.
- **User Distribution**: Subscribers form the bulk of the user base, but customers contribute significantly to longer trip durations.

These findings will be illustrated through well-designed visualizations to emphasize the key patterns and insights.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv("201902-fordgobike-tripdata.csv")
```

## Visualization 1:**Distribution of Trip Durations for User Types** This visualization
> It will display the distribution of trip durations separately for `Subscribers` and `Customers` to highlight differences in trip behaviors between these user groups.

```python
# Explanatory Visualization 1: Distribution of Trip Durations for User Types
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='duration_sec', hue='user_type', bins=30, kde=True, stat='density', palette='viridis')
plt.title('Distribution of Trip Durations for Subscribers and Customers', fontsize=14)
plt.xlabel('Trip Duration (seconds)', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.xlim(0, 5000)  # Focus on durations under 5000 seconds for better clarity
plt.legend(title='User Type', labels=['Customer', 'Subscriber'])
plt.grid(True)
plt.show()
```

This plot reveals the differences in trip duration distributions between subscribers and customers:

1. **Peak and Shape**:
   - Subscribers exhibit a much higher density of shorter trip durations, with a sharp peak under 500 seconds.
   - Customers also show a high density at shorter durations but have a broader spread, indicating a higher likelihood of longer trips compared to subscribers.

2. **Duration Focus**:
   - Most trips for both user types occur within the first 1000 seconds, aligning with typical short urban travel durations.
   - Customers demonstrate a notable presence of longer trip durations compared to subscribers, as observed from the broader tail.

3. **Density Distribution**:
   - The density of trips decreases sharply for both user types beyond 2000 seconds, emphasizing that extended trips are relatively rare.

This visualization effectively highlights the behavioral differences between user types, showing that subscribers predominantly engage in shorter trips, likely due to frequent and regular use. In contrast, customers show a more diverse pattern, possibly reflecting less frequent but longer-duration rides.

## Visualization 2: **Gender Distribution Among Subscribers and Customers**
> This visualization will display the gender distribution within the `Subscribers` and `Customers` user groups, highlighting any differences in gender composition between these two user types.

```python
# Explanatory Visualization 2: Gender Distribution Among Subscribers and Customers
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='user_type', hue='member_gender', palette='viridis')

# Adding plot details
plt.title('Gender Distribution Among Subscribers and Customers', fontsize=14)
plt.xlabel('User Type', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.legend(title='Gender', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

The visualization reveals distinct gender distribution patterns across the two user types: 

1. **Subscribers**:
   - The majority of `Subscribers` are **Male**, significantly outnumbering both **Female** and **Other** gender categories.
   - The count of `Female` subscribers is also substantial, but much lower than their male counterparts.
   - The `Other` gender category constitutes only a negligible proportion of the `Subscribers`.

2. **Customers**:
   - Similarly, among `Customers`, **Male** users represent the largest group, followed by **Female** users.
   - The difference between **Male** and **Female** counts is less pronounced for `Customers` compared to `Subscribers`.
   - The `Other` gender group is almost negligible among `Customers`, as well.

### Observations:
- `Male` users dominate both user types (`Subscribers` and `Customers`), but the ratio of `Female` users is relatively higher among `Customers` than among `Subscribers`.
- These patterns might reflect differences in preferences or accessibility to the bike-sharing system across genders within these user types.

This information could be useful for tailoring marketing strategies or improving service accessibility for underrepresented groups.

## Visualization 3: **Relationship Between Trip Duration, User Type, and Gender**
> This visualization will examine the relationship between `Trip Duration`, `User Type`, and `Gender`. By creating a box plot, we can highlight differences in trip duration across genders for both `Subscribers` and `Customers`, showcasing variations in usage patterns and preferences.

```python
# Explanatory Visualization 3: Relationship Between Trip Duration, User Type, and Gender
plt.figure(figsize=(10, 6))
sns.boxplot(
    data=df, 
    x='user_type', 
    y='duration_sec', 
    hue='member_gender', 
    palette='viridis'
)

# Adding plot details
plt.title('Relationship Between Trip Duration, User Type, and Gender', fontsize=14)
plt.xlabel('User Type', fontsize=12)
plt.ylabel('Trip Duration (seconds)', fontsize=12)
plt.ylim(0, 5000)  # Focus on trips under 5000 seconds for clarity
plt.legend(title='Gender', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

This box plot provides insights into the variation in trip duration across different user types (`Customer` and `Subscriber`) and genders (`Male`, `Female`, and `Other`). Below are the key observations:

- **Subscribers vs. Customers**: 
  - Customers generally have longer trip durations compared to Subscribers. The median trip duration for Customers is visibly higher across all gender groups.
  - The interquartile range (IQR) for Customers is also wider, indicating more variability in trip durations for this user type.

- **Gender Differences**:
  - Across both user types, Female users tend to have slightly longer trip durations on average compared to Male users.
  - Users in the `Other` gender category also show similar trip duration patterns to Female users, though the sample size for this category appears smaller.

- **Outliers**:
  - There are significant outliers in trip duration for all gender groups and user types, indicating that a small number of users take much longer trips compared to the majority. These outliers are more pronounced for Customers than Subscribers.

- **General Trends**:
  - Subscribers show more consistency in trip durations across genders, as evidenced by smaller IQRs and fewer extreme outliers.
  - Customers exhibit greater variability in trip durations, with a noticeable number of longer trips.

This visualization effectively highlights how user type and gender interact to influence trip duration, providing actionable insights into user behavior.
